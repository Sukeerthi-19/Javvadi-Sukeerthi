```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Read the CSV file
df = pd.read_csv(r'/content/PoliceKillingsUS.csv')

def analyze_categorical_association(var1, var2):
    """
    Analyze association between two categorical variables using:
    1. Cross-tabulation
    2. Chi-square test
    3. Visualization with heatmap
    """
    # Create contingency table
    cont_table = pd.crosstab(df[var1], df[var2])

    # Perform chi-square test
    chi2, p_value, dof, expected = chi2_contingency(cont_table)

    # Create heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(cont_table, annot=True, fmt='d', cmap='YlOrRd')
    plt.title(f'Association between {var1} and {var2}')
    plt.tight_layout()

    print(f"\nContingency Table:\n{cont_table}")
    print(f"\nChi-square statistic: {chi2:.2f}")
    print(f"p-value: {p_value:.4f}")
    print(f"Degrees of freedom: {dof}")

    if p_value < 0.05:
        print("\nThere is a significant association between the variables (p < 0.05)")
    else:
        print("\nThere is no significant association between the variables (p >= 0.05)")

# Example usage - analyzing association between 'race' and 'armed'
analyze_categorical_association('race', 'armed')

# Example usage - analyzing association between 'gender' and 'manner_of_death'
analyze_categorical_association('gender', 'manner_of_death')

plt.show()
```
