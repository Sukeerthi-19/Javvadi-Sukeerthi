```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
df = pd.read_csv('/mnt/data/cleaned_merged_heart_dataset.csv')

# Calculate mean
mean = df.mean()

# Calculate mode
mode = df.mode().iloc[0]  # .iloc[0] extracts the first mode value

# Calculate median
median = df.median()

# Calculate standard deviation
std_dev = df.std()

# Calculate variance
variance = df.var()

# Calculate range (max - min)
range_values = df.max() - df.min()

# Calculate skewness
skewness = df.skew()

# Identify outliers using Z-score (values with Z-score > 3 or < -3)
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))  # Numeric columns only
outliers = (z_scores > 3).any(axis=1)  # Flag rows with Z-score > 3 in any column

# Collecting outlier rows
outlier_rows = df[outliers]

# Print results
print("### Statistical Summary ###\n")

print("Mean:\n", mean)
print("\nMode:\n", mode)
print("\nMedian:\n", median)
print("\nStandard Deviation:\n", std_dev)
print("\nVariance:\n", variance)
print("\nRange:\n", range_values)
print("\nSkewness:\n", skewness)

print("\n### Outliers ###\n")
print(outlier_rows.head())  # Display the first few outlier rows
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
df = pd.read_csv('/content/Credit_Card_Applications.csv')

# Calculate mean for numeric columns
mean = df.mean(numeric_only=True)

# Calculate mode
mode = df.mode().iloc[0]

# Calculate median for numeric columns
median = df.median(numeric_only=True)

# Calculate standard deviation for numeric columns
std_dev = df.std(numeric_only=True)

# Calculate variance for numeric columns
variance = df.var(numeric_only=True)

# Calculate range (max - min) for numeric columns
range_values = df.max(numeric_only=True) - df.min(numeric_only=True)

# Calculate skewness for numeric columns
skewness = df.skew(numeric_only=True)

# Identify outliers using Z-score (Z > 3 or Z < -3)
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
outliers = (z_scores > 3).any(axis=1)

# Collecting outlier rows
outlier_rows = df[outliers]

# Display results
print("### Statistical Summary ###\n")
print("Mean:\n", mean)
print("\nMode:\n", mode)
print("\nMedian:\n", median)
print("\nStandard Deviation:\n", std_dev)
print("\nVariance:\n", variance)
print("\nRange:\n", range_values)
print("\nSkewness:\n", skewness)

print("\n### Outliers (First 5 Rows) ###\n")
print(outlier_rows.head())
```
