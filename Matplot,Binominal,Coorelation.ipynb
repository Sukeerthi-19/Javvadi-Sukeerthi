Matplot


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (update the path as needed)
data = pd.read_csv('Credit_Card_Applications.csv')

# Ensure the plotting style is compatible
plt.style.use('ggplot')

# Plot 1: Histogram of A2 (Income)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data['A2'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of A2 (Income)')
plt.xlabel('Income')
plt.ylabel('Frequency')

# Plot 2: Scatter plot of A2 (Income) vs. A3 (Age) by Class
plt.subplot(1, 2, 2)
for class_value, color in zip([0, 1], ['red', 'green']):
    subset = data[data['Class'] == class_value]
    plt.scatter(subset['A2'], subset['A3'], alpha=0.6, label=f'Class {class_value}', color=color)

plt.title('Income vs. Age by Class')
plt.xlabel('A2 (Income)')
plt.ylabel('A3 (Age)')
plt.legend(title='Class', loc='upper left')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

np.random.binominal()

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (update the path as needed)
data = pd.read_csv('Credit_Card_Applications.csv')

# Generate random binomial data
# n: number of trials, p: probability of success, size: number of samples to match dataset
n, p = 10, 0.5  # 10 trials, 50% probability of success
data['RandomBinomial'] = np.random.binomial(n, p, size=len(data))

# Plotting with Seaborn
plt.figure(figsize=(12, 6))

# Distribution plot comparing A3 (Age) and Random Binomial data
sns.histplot(data['A3'], color='blue', kde=True, label='A3 (Age)', bins=20)
sns.histplot(data['RandomBinomial'], color='orange', kde=True, label='Random Binomial', bins=20)

plt.title('Comparison of A3 (Age) Distribution and Random Binomial Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (update the path as needed)
data = pd.read_csv('Credit_Card_Applications.csv')

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

plt.title('Correlation Matrix Heatmap')
plt.show()
```
